# **Import Necessary Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Analysis of ODI Players**

```python
df = pd.read_csv('/content/ODI data.csv')
```

```python
df.head()
```

# **Players with More than 15000 ODI Runs**

```python
df['Runs'] = pd.to_numeric(df['Runs'], errors='coerce')
df.query('Runs > 15000')
```

# **Top 10 Run Scorers in ODI**

```python
top_10_scorers = df.groupby('Player')['Runs'].sum().sort_values(ascending=False).head(10)
top_10_scorers
```

## **Visualize Top 10 Run Scorers**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_scorers.index, y=top_10_scorers.values)
plt.title('Top 10 Run Scorers in ODIs')
plt.xticks(rotation=45)
plt.show()
```

# **Analysis of Top Test Bowlers**

```python
bowl_test = pd.read_csv('/content/Bowling_test.csv')
```

```python
bowl_test.head()
```

# **Top Test Wicket Takers**

```python
# Remove duplicates based on the 'Player' column, keeping the first occurrence.
bowl_test_no_duplicates = bowl_test.drop_duplicates(subset='Player', keep='first')

# Sort the DataFrame by the 'Wkts' column in descending order to get the top wicket-takers.
top_wicket_takers = bowl_test_no_duplicates.sort_values(by='Wkts', ascending=False)

# Display the top wicket-takers along with a new column showing their rank
top_wicket_takers['top_wkts'] = range(1, len(top_wicket_takers) + 1)
top_wicket_takers
```

# **Visualize Top Test Wicket Takers**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_wicket_takers.index, y=top_wicket_takers['Wkts'])
plt.title('Top 10 wicket takers in test')
plt.xticks(rotation=45)
plt.show()
```
